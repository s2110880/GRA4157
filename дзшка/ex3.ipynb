## Exercise 3: Scrape Wikipedia's List of Best-selling Music Artists


1. **Initial Scrape:** Scrape Wikipedia's table of best-selling music artists (https://en.wikipedia.org/wiki/List_of_best-selling_music_artists).


```python
#impports
import requests
from bs4 import BeautifulSoup
```

```python
result = requests.get("https://en.wikipedia.org/wiki/List_of_best-selling_music_artists")
```

```python
#extracting html source
src = result.content
document = BeautifulSoup(src, 'lxml')
```

```python
#Wikipedia tables all have  class wikitable (there are 6 table on the page)
table = document.find_all("table", {"class": "wikitable"})
len(table)
```

```python
# since there are 6 separate tables we can specify which table's rows we print (example for table 1)
rows = table[0].find_all("tr")

# going through rows and extracting the cells
for row in rows:
    cells = row.find_all(["td", "th"])
    cells_text = [cell.get_text(strip=True) for cell in cells]
    
    print(cells_text)
```

```python
for i in range(6):  #loop through 6 tables
    rows = table[i].find_all("tr")  #getting rows from tables 
    #going through rows and printing data cells
    for row in rows:
        cells = row.find_all(["td", "th"])
        cells_text = [cell.get_text(strip=True) for cell in cells]
        print(cells_text)
    #just spacing between tables
    print("\n") 
```

2. **Data Cleanup:** Retain only 'Artist', 'Country/Market', and 'Certified Sales'.

```python
for i in range(6):  # Loop through the first 6 tables    
    rows = table[i].find_all("tr")
    for row in rows:
        cells = row.find_all(["td", "th"])
        cells_text = [cell.get_text(strip=True) for cell in cells]
        
         
        artist = cells_text[0]  
        country_market = cells_text[1]  
        certified_sales = cells_text[5].split('US')[0].split('FRA')[0].split('JPN')[0].split('UK')[0] #to filter everthing that goes after total sum
        print(artist, country_market, certified_sales)
```

3. **Analysis:** Find the artist with the highest certified sales.

```python
artist_sales_data = [] #empty list for third part of the Ex.
for i in range(6):  # Loop through the first 6 tables    
    rows = table[i].find_all("tr")
    for row in rows:
        cells = row.find_all(["td", "th"])
        cells_text = [cell.get_text(strip=True) for cell in cells]
        
         
        artist = cells_text[0]  
        country_market = cells_text[1]  
        certified_sales = cells_text[5].split('US')[0].split('FRA')[0].split('JPN')[0].split('UK')[0] #to filter everthing that goes after total sum
        if "million" in certified_sales: #because of the lines like "Artist Country Total certified units(from available markets)"
            sales_number = float(certified_sales.replace("million", "").strip())
        artist_sales_data.append((artist, country_market, sales_number))
        print(artist, country_market, sales_number)
```

```python
#creating a loop that finds the maximum sales_number
artist_with_highest_sales = ""
highest_sales = 0
for artist, country_market, sales_number in artist_sales_data:
    if sales_number > highest_sales:
        highest_sales = sales_number
        artist_with_highest_sales = artist
print(artist_with_highest_sales, highest_sales)
```
